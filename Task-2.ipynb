```python
import tkinter as tk
from math import *

def on_click(event):
    current_text = entry.get()
    text = event.widget.cget("text")
    if text == "=":
        try:
            result = eval(current_text)
            entry.delete(0, tk.END)
            entry.insert(tk.END, str(result))
        except Exception as e:
            entry.delete(0, tk.END)
            entry.insert(tk.END, "Error")
    elif text == "C":
        entry.delete(0, tk.END)
    elif text == "√":
        try:
            current_text = eval(current_text)
            result = sqrt(current_text)
            entry.delete(0, tk.END)
            entry.insert(tk.END, str(result))
        except Exception as e:
            entry.delete(0, tk.END)
            entry.insert(tk.END, "Error")  
    elif text == "π":
        entry.insert(tk.END, str(3.1415))
    else:
        entry.insert(tk.END, text)

def create_button(text, row, col, col_span=1):
    button = tk.Button(root, text=text, font=("Helvetica", 18), bd=5)
    button.grid(row=row, column=col, columnspan=col_span, padx=5, pady=5)
    button.bind("<Button-1>", on_click) 
    return button

root = tk.Tk()
root.title("CodSoft's Python Programming Task-2")
root.geometry("350x470")

heading_label = tk.Label(root, text="Calculator", font=("Times New Roman", 25, "bold"), highlightcolor=("red"))
heading_label.grid(row=0, column=0, columnspan=4, pady=20)

entry = tk.Entry(root, font=("Times New Roman", 24), bd=5, justify=tk.RIGHT)
entry.grid(row=1, column=0, columnspan=4, padx=10, pady=10)

create_button("π", 2, 0)
create_button("√", 2, 1)
create_button("%", 2, 2)
create_button("C", 2, 3)

create_button("7", 3, 0)
create_button("8", 3, 1)
create_button("9", 3, 2)
create_button("+", 3, 3)

create_button("4", 4, 0)
create_button("5", 4, 1)
create_button("6", 4, 2)
create_button("-", 4, 3)

create_button("1", 5, 0)
create_button("2", 5, 1)
create_button("3", 5, 2)
create_button("*", 5, 3)

create_button(".", 6, 0)
create_button("0", 6, 1)
create_button("=", 6, 2)
create_button("/", 6, 3)

root.mainloop()
```
